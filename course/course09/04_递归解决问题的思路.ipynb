# Tower of Hanoi
本章中，我们将通过实现古老的河塔游戏来感受递归解决问题的能力  
下面是Hanoi游戏的截图，游戏的目标是把第一根棍子上的tower移动到第二根棍子上

```python
def move_disk(dis_number, from_peg, to_peg):
    print(
        "Move disk "
        + str(dis_number)
        + " from peg "
        + str(from_peg)
        + " to "
        + str(to_peg)
    )


def solve_hanoi(n, start_peg, end_peg):
    if n == 1:
        move_disk(n, start_peg, end_peg)
    else:
        spare_peg = 6 - start_peg - end_peg
        solve_hanoi(n - 1, start_peg, spare_peg)
        move_disk(n, start_peg, end_peg)
        solve_hanoi(n - 1, spare_peg, end_peg)


solve_hanoi(4, 1, 2)
```

上面递归算法的时间复杂度为 n^2
