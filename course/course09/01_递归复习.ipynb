# 递归复习
递归是一种函数循环的模式，虽然没那么直观，但是写起来优美

下面有两个函数，都是阶乘函数的实现，其中一个使用while loop。  
在下代码中,实现的逻辑在于保存一个结果符合预期的变量，然后返回  
运行过程如下  
![Alt text](image.png)  
可以看到，出现了4次运算，计算了4次n阶乘

```python
def factorial(n):
    fact = 1
    i = 1
    while i <= n:
        fact *= i
        i += 1
    return fact


factorial(10)
```

另一个则是使用递归来实现的,可以看到结果相同，但是实现逻辑不同
在递归中，不需要考虑中间的逻辑，只需要考虑两件事：
- 平常的情况：在这个例子中是 !n = n * !(n - 1)
- 结束的条件：在这个例子中是 !0 = 1  

运行过程如下图  
  
运行效率更高，一次完成，没有任何中间步骤的浪费  

```python
def factorial_recursive(n):
    if n == 0:
        return 1
    else:
        return factorial_recursive(n - 1) * n


factorial_recursive(10)
```

总结一下， 和之前的循环不同，循环是在求符合逻辑的返回值  
而递归则是在构建一个expression,并且逐级递归，直到结束条件出现。  
例如：!n = !(n - 1) * n。  
两者功能类似，有时候使用递归更加优雅，有时候使用循环结构更加好，**没有绝对的对错**。
